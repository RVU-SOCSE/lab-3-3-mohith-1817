```python
#without using update()
ph={
    'A':1234,
    'B':5698,
    'C':789
}
print(ph)
print("After updating")
ph["C"]=888
print(ph)
```

```python
#values any data type int, float, string , list, etc
ph={
    'A':1234,
    'B':5698,
    'C':789
}
print(ph)
ph["C"]={888,999,777}      #list allowed
print(ph)
ph["B"]={"B-home":7145585,"B-office":4569885}  # dictionary

print(ph)

print(ph["B"]["B-home"])
```
